```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

# Matplotlib

```python
x_sepal = df['SepalLengthCm']
y_sepal= df['SepalWidthCm']

x_petal = df['PetalLengthCm']
y_petal = df['PetalWidthCm']

fig , ax = plt.subplots(2 , 1 , figsize=(10,10))
ax[0].scatter(x_sepal,y_sepal , )
ax[1].scatter(x_petal,y_petal)
ax[0].legend()
ax[0].set_title('Sepal ')
ax[0].set_xlabel('Sepal longeur')
ax[0].set_ylabel('Sepal largeur')
ax[1].legend()
ax[1].set_title('Petal')
ax[1].set_xlabel('Sepal longeur')
ax[1].set_ylabel('Sepal largeur')
plt.show()
```

# Searborn

```python
sns.set_style('whitegrid')
fig , ax = plt.subplots(2 , 1 , figsize=(10,10))
sns.scatterplot(data=df , x='SepalLengthCm' , y='SepalWidthCm' , hue='Species' , ax=ax[0])
sns.scatterplot(data=df , x='PetalLengthCm' , y='PetalWidthCm' , hue='Species' , ax=ax[1])
```
